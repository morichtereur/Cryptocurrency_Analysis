```python
# Visualize the close price over time
ax = df.plot(x='Date', y='Close')
ax.set_xlabel("Date")
ax.set_ylabel("Close Price (USD)")
ax.set_title("Close Price of S&P500")
```

# Examine the volatility for the year 2023

```python
# Filter the dates for the year 2020
start_date = "2023-01-01"
end_date = "2023-12-31"

after_start_date = df["Date"] >= start_date
before_end_date = df["Date"] <= end_date
between_two_dates = after_start_date & before_end_date
year_2023 = df.loc[between_two_dates]

print(year_2023)
```

```python
# Create a new dataframe showing only 2020
df_2023 = pd.DataFrame(data=year_2023)
df_2023.shape
df_2023.head()
```

```python
# Visualize the S&P500 close price for 2023
ax = df_2023.plot(x='Date', y='Close');
ax.set_xlabel("Date")
ax.set_ylabel("Close Price (USD)")
ax.set_title("Close Price of S&P500 in 2023")
```

```python
# Add a column called "Log_Returns" with the daily log return of the Close price
df_2023["Log_Returns"] = np.log(df_2023['Close']/df_2023['Close'].shift())
df_2023.head()
```

```python
# We need the std deviation for the volatility of the stock
# df_2023["Log_Returns"].std()

# Volatility is defined as the annualized standard deviation. There are ~253 trading days per year. Square root = **.5
volatility_2023 = df_2023["Log_Returns"].std()*253**.5
```

```python
# Calculate the volatility of the 2023 stock price of S&P500
str_vol_2023 = str(round(volatility_2023, 3)*100)

print(str_vol_2023)
```

```python
# Visualize the volatility of the 2023 stock price of S&P500

fig, ax = plt.subplots()
df_2023["Log_Returns"].hist(ax=ax, bins=50, alpha=0.6, color="b")
ax.set_xlabel("Log Return")
ax.set_ylabel("Frequency of Log Return")
ax.set_title("2023 S&P500 Volatility: " + str_vol_2023 + "%" )
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

# Sample DataFrames, ensure your Date column is datetime type
df_2021['Date'] = pd.to_datetime(df_2021['Date'])
df_2022['Date'] = pd.to_datetime(df_2022['Date'])
df_2023['Date'] = pd.to_datetime(df_2023['Date'])

x1 = df_2021["Date"]
x2 = df_2022["Date"]
x3 = df_2023["Date"]

a = df_2021["Close"].values
b = df_2022["Close"].values
c = df_2023["Close"].values

plt.figure(figsize=(10, 5))  # Set a figure size if needed
plt.plot(x1, a, label="2021")
plt.plot(x2, b, label="2022")
plt.plot(x3, c, label="2023")

plt.title("Close Price of S&P500 (2021-2023)")
plt.xlabel("Date")
plt.ylabel("Close Price (USD)")

# Format the dates on the x-axis
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Locate ticks at every year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # Format ticks as 'Year-Month-Day'

plt.gcf().autofmt_xdate()  # Rotate date labels to prevent overlap

plt.legend()
plt.show()
```

```python
# Visualize the volatility percent as a bar chart
# plt.bar(x, height, width, bottom, align)

year = ["2021", "2022", "2023"]
values = [float(str_vol_2021), float(str_vol_2022), float(str_vol_2023)]

colors = ["blue", "orange", "green"]

fig = plt.figure(figsize = (8, 5))
 
# creating the bar plot
plt.bar(x=year, height=values, color=colors, edgecolor='black', width = 0.4, bottom=0)
 
plt.xlabel("Year")
plt.ylabel("Percent Volatility")
plt.title("Annual Percent Volatility of S&P500 (2021-2023) ")
plt.show()
```
