```python
data = [
    ("Binance Coin", 139.0, 73.6, 43.9),  # Values from your list for Binance Coin
    ("Bitcoin", 80.0, 65.0, 43.37),       # Values from your list for Bitcoin
    ("Cardano", 126.0, 92.3, 66.2),       # Values from your list for Cardano
    ("Ethereum", 106.0, 87.7, 45.9),      # Values from your list for Ethereum
    ("Polkadot", 145.0, 93.7, 62.9),      # Values from your list for Polkadot
    ("Ripple", 145.0, 85.6, 78.91),       # Values from your list for Ripple
    ("Solana", 161.0, 132.0, 95.2),       # Values from your list for Solana
    ("S&P 500", 13.1, 24.3, 13.1),       
]
```

```python
# Create summary dataframe of the annual volatility
df = pd.DataFrame(data=data, columns=['coin_name', 'year_2021', 'year_2022', 'year_2023'])
df
```

```python
import matplotlib.pyplot as plt

# Data and years for the graph
year = ["2021", "2022", "2023"]
x = year

# Updated data for each coin
a = (139.0, 73.6, 43.9)  # Binance Coin
b = (80.0, 65.0, 43.37)  # Bitcoin
c = (126.0, 92.3, 66.2)  # Cardano
d = (106.0, 87.7, 45.9)  # Ethereum
e = (145.0, 93.7, 62.9)  # Polkadot
f = (145.0, 85.6, 78.91)  # Ripple
g = (161.0, 132.0, 95.2)  # Solana
h = (13.1, 24.3, 13.1)   # S&P 500

# Creating the plot
plt.figure(figsize=(14, 8))

plt.plot(x, a, marker='o', linestyle='-', linewidth=2, label="Binance Coin")
plt.plot(x, b, marker='s', linestyle='--', linewidth=2, label="Bitcoin")
plt.plot(x, c, marker='^', linestyle='-.', linewidth=2, label="Cardano")
plt.plot(x, d, marker='>', linestyle=':', linewidth=2, label="Ethereum")
plt.plot(x, e, marker='p', linestyle='-', linewidth=2, label="Polkadot")
plt.plot(x, f, marker='*', linestyle='--', linewidth=2, label="Ripple")
plt.plot(x, g, marker='D', linestyle='-.', linewidth=2, label="Solana")
plt.plot(x, h, marker='x', linestyle=':', linewidth=2, label="S&P 500")

# Setting up the x and y ticks, labels, and title
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Year", fontsize=16)
plt.ylabel("Annual Percent Volatility", fontsize=16)
plt.title("Annual Percent Volatility (2021-2023)", fontsize=20)
plt.grid(True)

# Adding a legend
plt.legend(fontsize=12, loc='upper right', bbox_to_anchor=(1.15, 1))

# Saving the figure
plt.savefig("ImprovedAnnualPercentVolatilityAllCoins2021-2023.png", bbox_inches='tight')

# Display the plot
plt.show()
```

## Most and Least Volatile Each Year:

### 2021:
- **Most Volatile**: Solana (161.0%)

### 2023:
- **Most Volatile**: Solana (95.2%)
- **Least Volatile**: S&P 500 (13.1%)

## Overall (2021-2023):
To determine overall volatility, the range (difference between the highest and lowest volatility values) for each coin and the S&P 500 across the three years was analyzed. A higher range indicates greater fluctuations in volatility.

### Volatility Ranges:
- **Binance Coin**: 95.1% (139.0 - 43.9)
- **Bitcoin**: 36.63% (80.0 - 43.37)
- **Cardano**: 59.8% (126.0 - 66.2)
- **Ethereum**: 60.1% (106.0 - 45.9)
- **Polkadot**: 82.1% (145.0 - 62.9)
- **Ripple (XRP)**: 79.09% (145.0 - 65.91)
- **Solana**: 65.8% (161.0 - 95.2)
- **S&P 500**: 11.2% (24.3 - 13.1)

### Most and Least Volatile Over Time:
- **Most Volatile Over Time**: Solana (range: 65.8%), though Binance Coin and Polkadot also show significant volatility.
- **Least Volatile Over Time**: S&P 500 (range: 11.2%), followed by Bitcoin (range: 36.63%).

## Summary:
- **Solana** was consistently the most volatile each year, underlining its higher risk and potential for significant price changes. This makes it a key consideration for high-risk tolerance investors.
- **Bitcoin** and the **S&P 500** maintained relatively low volatility levels, indicating more stability compared to other cryptocurrencies.
- The **S&P 500**, as a broad market index, exhibited considerably lower volatility than individual cryptocurrencies, reflecting its diverse asset composition which mitigates the impact of volatility from any single asset.
