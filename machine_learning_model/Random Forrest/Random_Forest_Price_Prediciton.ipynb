```python
# Import the datetime module and alias it as dt
import datetime as dt

# Import the pandas library and alias it as pd
import pandas as pd

# Import the numpy library and alias it as np
import numpy as np

# Import the arange function from the numpy module
from numpy import arange

# Import the pyplot module from the matplotlib library and alias it as plt
import matplotlib.pyplot as plt

# Import the read_csv function from the pandas module
from pandas import read_csv

# Import the metrics module from the sklearn library
from sklearn import metrics

# Import the train_test_split function from the model_selection module in sklearn
from sklearn.model_selection import train_test_split

# Import the RandomForestRegressor class from the ensemble module in sklearn
from sklearn.ensemble import RandomForestRegressor

# Import the StandardScaler class from the preprocessing module in sklearn
from sklearn.preprocessing import StandardScaler

# Import the RandomizedSearchCV class from the model_selection module in sklearn
from sklearn.model_selection import RandomizedSearchCV
```

```python
import pandas as pd

# Specify the file path
file_path = '/Users/moritzrichter/Desktop/cryptocurrency_analysis/data/Bitcoin_(BTC)/bitcoin(fiveyears).csv'

# Load the dataset from the specified file path
bitcoin_df = pd.read_csv(file_path)

# Convert 'Date' to datetime to ensure it sorts and indexes correctly
bitcoin_df['Date'] = pd.to_datetime(bitcoin_df['Date'])

# Sort the DataFrame by the 'Date' column in ascending order (oldest first)
bitcoin_df.sort_values('Date', ascending=True, inplace=True)

# Set 'Date' as the index of the DataFrame
bitcoin_df.set_index('Date', inplace=True)

# Display the first few rows of the DataFrame to ensure it has loaded and sorted correctly
print(bitcoin_df.head())
```

```python
# Drop rows with missing values
bitcoin_df.dropna(inplace=True)

# Drop the 'Name' column
bitcoin_df.drop(columns=['Name'], inplace=True)

# Display the first few rows of the DataFrame after dropping NaN rows and the 'Name' column
print(bitcoin_df.head())
```

```python
y = bitcoin_df.Close
x = bitcoin_df.drop(columns=["Close"], axis=1)
```

```python
# Split and train Data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.26,  random_state=0)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(x_train)

# Scale the data
X_train_scaled = X_scaler.transform(x_train)
X_test_scaled = X_scaler.transform(x_test)
```

```python
model = RandomForestRegressor(n_estimators=700, random_state=42, min_samples_split=2, min_samples_leaf=1, max_depth=10, bootstrap=True)
model.fit(x_train, y_train)
predict = model.predict(x_test)
print(predict)
print(predict.shape)
```

```python
print(predict.shape)
```

```python
print("Mean Absolute Error:", round(metrics.mean_absolute_error(y_test, predict), 4))
print("Mean Squared Error:", round(metrics.mean_squared_error(y_test, predict), 4))
print("Root Mean Squared Error:", round(np.sqrt(metrics.mean_squared_error(y_test, predict)), 4))
print("(R^2) Score:", round(metrics.r2_score(y_test, predict), 4))
print(f'Train Score : {model.score(x_train, y_train) * 100:.2f}% and Test Score : {model.score(x_test, y_test) * 100:.2f}% using Random Tree Regressor.')
errors = abs(predict - y_test)
mape = 100 * (errors / y_test)
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.') 
```

```python
import pandas as pd

# Generate predictions DataFrame with a period of one year
predictions = pd.DataFrame({"Predictions": predict}, index=pd.date_range(start=bitcoin_df.index[-1], periods=len(predict), freq="D"))
predictions.to_csv("Predicted-price-data.csv")

# Collect future days from predicted values for one year
oneyear_df = pd.DataFrame(predictions[:252])
oneyear_df.to_csv("one-year-predictions.csv")

# Collect future days from predicted values for one month
onemonth_df = pd.DataFrame(predictions[:21])
onemonth_df.to_csv("one-month-predictions.csv")

# Collect future days from predicted values for five days
fivedays_df = pd.DataFrame(predictions[:5])
fivedays_df.to_csv("five-days-predictions.csv")
```
